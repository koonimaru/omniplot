```python
import sys, os
import sys

# sys.path.append("../omniplot")
from omniplot import plot as op
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
%config InlineBackend.print_figure_kwargs={'bbox_inches': None}
```

```python
df = sns.load_dataset("penguins")
df = df.dropna(axis=0)
# features=["species","sex","bill_length_mm","bill_depth_mm","flipper_length_mm","body_mass_g"]
# df=df[features]
print(df.shape)
op.scatterplot(
    df,
    x="bill_length_mm",
    y="bill_depth_mm",
    #    category=['species',"sex"],
    colors=["flipper_length_mm"],
    sizes="body_mass_g",
    show_labels={"val": "body_mass_g", "topn": 5},
)
```

```python
op.scatterplot(
    df,
    x="bill_length_mm",
    y="bill_depth_mm",
    palette_cat="tab20b",
    category=["species", "sex"],
    colors=["flipper_length_mm"],
    markers=True,
)
```

```python
# features=["species","sex","bill_length_mm","bill_depth_mm","flipper_length_mm","body_mass_g"]
# df=df[features]
print(df.shape)
op.scatterplot(
    df,
    x="bill_length_mm",
    y="bill_depth_mm",
    xunit="mm",
    yunit="mm",
    category=["species", "sex", "island"],
    colors=["flipper_length_mm"],
    sizes="body_mass_g",
    show_labels={"val": "body_mass_g", "topn": 5},
    marginal_dist=True,
)
```

```python
op.scatterplot(
    df,
    x="bill_length_mm",
    y="bill_depth_mm",
    xunit="mm",
    yunit="mm",
    category=["species", "sex", "island"],
    colors=["flipper_length_mm"],
    sizes="body_mass_g",
    show_labels={"val": "body_mass_g", "topn": 5},
    marginal_dist=True,
    kmeans=True,
    save="scatter_kmeans.pdf",
)
```

```python
op.scatterplot(
    df,
    x="bill_length_mm",
    y="bill_depth_mm",
    xunit="mm",
    yunit="mm",
    category=["species", "sex", "island"],
    colors=["flipper_length_mm"],
    sizes="body_mass_g",
    show_labels={"val": "body_mass_g", "topn": 5},
    marginal_dist=True,
    regression=True,
)
```

```python
op.scatterplot(df, x="bill_length_mm", y="bill_depth_mm", xunit="mm", yunit="mm")
```
