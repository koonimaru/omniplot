```python
import seaborn as sns
from omniplot import plot as op
import numpy as np
import pandas as pd
```

```python
f="../data/energy_vs_gdp.csv"
df=pd.read_csv(f, comment='#')
df=df.set_index("country")
```

```python
# data contains countries' electricity usages by resources and population, GDP per capita in 2018.
df
```

```python
op.pie_scatter(df, x="gdppc",y="pop", category=['biofuel_electricity',
                                             'coal_electricity',
                                             'gas_electricity',
                                             'hydro_electricity',
                                             'nuclear_electricity',
                                             'oil_electricity',
                                             'other_renewable_electricity',
                                             'solar_electricity',
                                             'wind_electricity'],logscalex=True,logscaley=True,piesize=0.01)
```

```python
# resizing pie charts according to total electricity usages
op.pie_scatter(df, x="gdppc",y="pop", category=['biofuel_electricity',
                                             'coal_electricity',
                                             'gas_electricity',
                                             'hydro_electricity',
                                             'nuclear_electricity',
                                             'oil_electricity',
                                             'other_renewable_electricity',
                                             'solar_electricity',
                                             'wind_electricity'],logscalex=True,logscaley=True,
                                            sizes="sum_of_each",
                                            min_piesize=0.1,piesize=0.05)
```

```python
op.pie_scatter(df, x="gdppc",y="pop", category=['biofuel_electricity',
                                             'coal_electricity',
                                             'gas_electricity',
                                             'hydro_electricity',
                                             'nuclear_electricity',
                                             'oil_electricity',
                                             'other_renewable_electricity',
                                             'solar_electricity',
                                             'wind_electricity'],logscalex=True,logscaley=True,
                                            sizes="sum_of_each",label="topn_of_sum",
                                            min_piesize=0.1,piesize=0.05)
```
