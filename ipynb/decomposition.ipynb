```python
import seaborn as sns
from omniplot import plot as op
```

```python
df = sns.load_dataset("penguins")
df = df.dropna(axis=0)
features = [
    "species",
    "bill_length_mm",
    "bill_depth_mm",
    "flipper_length_mm",
    "body_mass_g",
]
df = df[features]
res = op.decomplot(df, category="species", method="pca", markers=False)
```

```python
res = op.decomplot(df, category="species", method="nmf")
```
