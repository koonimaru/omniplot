`````python
f = "owid-energy-data.csv"
df = pd.read_csv(f)
_df = df.loc[df["country"] == "United States"]
cols = [
    "biofuel_consumption",
    "coal_consumption",
    "gas_consumption",
    "hydro_consumption",
    "nuclear_consumption",
    "oil_consumption",
    "other_renewable_consumption",
    "solar_consumption",
    "wind_consumption",
]
op.stackedlines(
    df=_df,
    x="year",
    y=cols,
    title="United States",
    ylabel="terawat per hour",
    remove_all_zero=True,
    show_values=True,
)
`````

`````python
op.stackedlines(
    df=_df,
    x="year",
    y=cols,
    title="United States",
    ylabel="terawat per hour",
    remove_all_zero=True,
    show_values=True,
    inverse=True,
)
`````

`````python
# show years with no record
op.stackedlines(df=_df, x="year", y=cols, title="United States", show_values=True)
`````

`````python
# Do not show percentages
op.stackedlines(
    df=_df,
    x="year",
    y=cols,
    title="United States",
    remove_all_zero=True,
    show_values=False,
)
`````

`````python
# without sorting
op.stackedlines(
    df=_df,
    x="year",
    y=cols,
    title="United States",
    remove_all_zero=True,
    show_values=False,
    sort=False,
)
`````

`````python
_df = _df.reset_index()
`````
