Drawing complex heatmap with clustering

````python
import seaborn as sns
from omniplot import plot as op
import pandas as pd
import numpy as np
````

in the case, you want to use GUI, put the following line in to the code block
```
%matplotlib tk 
```
for macos, "%matplotlib qt5" may work after installing pyqt5 with pip

````python
%config InlineBackend.print_figure_kwargs={'bbox_inches': None}
````

````python
df = sns.load_dataset("penguins")

df = df.dropna(axis=0)
# df=df.reset_index(drop=True)
res = op.heatmap(
    df=df,
    variables=["bill_length_mm", "bill_depth_mm", "flipper_length_mm"],
    row_split=True,
    clustering_method="hierarchical",
    category=["species", "island"],
    col_colors={"organ": ["bill", "bill", "flipper"]},
    row_plot=["body_mass_g"],
    row_ticklabels=False,
    approx_clusternum=3,
)
````

````python
res = op.heatmap(
    df=df,
    variables=["bill_length_mm", "bill_depth_mm", "flipper_length_mm"],
    row_split=True,
    clustering_method="kmeans",
    category=["species", "island"],
    col_colors={"organ": ["bill", "bill", "flipper"]},
    row_plot=["body_mass_g"],
    row_ticklabels=False,
    approx_clusternum=3,
)
````

````python
res = op.heatmap(
    df=df,
    dtype="categorical",
    variables=["species", "island", "sex"],
    row_plot=["body_mass_g"],
    row_ticklabels=False,
    approx_clusternum=3,
    column_wise_color=True,
)
````

````python
mat = np.concatenate(
    [
        0.25 * np.random.uniform(0, 1, size=[10, 25]),
        0.5 * np.random.uniform(0, 1, size=[15, 25]),
        np.random.uniform(0, 1, size=[15, 25]),
    ]
)
df = pd.DataFrame(mat)
res = op.heatmap(
    df,
    sizes=mat,
    row_split=True,
    row_plot={"normal": np.random.normal(size=mat.shape[0])},
    row_scatter={"uniform": np.random.uniform(size=mat.shape[0])},
    row_bar={"range": np.arange(mat.shape[0])},
    col_bar={"range": np.arange(mat.shape[1])},
    edgecolor=None,
    approx_clusternum=3,
    ztranform=False,
)
````

````python
res = op.heatmap(
    df,
    shape="circle",
    sizes=mat,
    row_split=True,
    row_plot={"normal": np.random.normal(size=mat.shape[0])},
    row_scatter={"uniform": np.random.uniform(size=mat.shape[0])},
    row_bar={"range": np.arange(mat.shape[0])},
    col_bar={"range": np.arange(mat.shape[1])},
    edgecolor=None,
    approx_clusternum=3,
    ztranform=False,
)
````

````python
res = op.heatmap(
    df,
    sizes=mat,
    size_title="random",
    cbar_title="random colors",
    # col_split=True,
    row_plot={"mean": mat.mean(axis=1)},
    row_scatter={"uniform": np.random.uniform(size=mat.shape[0])},
    row_axis=2,
    col_bar={"range": np.arange(mat.shape[1])},
    edgecolor=None,
    approx_clusternum=3,
    ztranform=False,
)

_mat = mat[res["rowsort"]]

ax = res["axes"]["row_0"]
op.violinplot2(_mat, ax=ax, orientation="horizontal")
ax.set_xlabel("violin")
ax.set_ylim(-0.5, mat.shape[0] - 0.5)

ax = res["axes"]["row_1"]
_df = pd.DataFrame(
    {"<0.5": np.sum(_mat < 0.5, axis=1), ">=0.5": np.sum(_mat >= 0.5, axis=1)}
)
op.stacked_barplot_num(
    df=_df, ax=ax, orientation="horizontal", show_values=False, show_legend=False
)
ax.invert_yaxis()
ax.set_xlabel("proportion")
ax.set_yticks([])
h, l = ax.get_legend_handles_labels()
res["axes"]["row_legend0"].legend(h, l, loc="upper right", bbox_to_anchor=[1.5, 1])
````

````python
# op.heatmap(df, shape="circle", sizes=mat, edgecolor=None, approx_clusternum=3, row_split=True, ztranform=False)
mat = np.arange(50).reshape([5, 10]).astype(np.float)
df = pd.DataFrame(mat)
op.heatmap(
    df,
    shape="triangle",
    sizes=mat,
    row_cluster=False,
    col_cluster=False,
    edgecolor=None,
    approx_clusternum=3,
    ztranform=False,
    row_split=True,
)
````
