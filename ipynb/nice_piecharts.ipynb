```python
import seaborn as sns
from omniplot import plot as op
import pandas as pd
```

```python
df = sns.load_dataset("penguins")
df = df.dropna(axis=0)
tmp = []
for num, (sp, i, se) in enumerate(zip(df["species"], df["island"], df["sex"])):
    if num / df.shape[0] > 0.98:
        tmp.append("NA")
    else:
        tmp.append(sp[0] + "_" + i[0] + "_" + se[0])
df["combine"] = tmp
# print(df)
res = op.nice_piechart(
    df,
    category=["species", "island", "sex", "species", "island", "sex", "combine"],
    ncols=4,
    show_legend=True,
)
```

```python
res = op.nice_piechart(
    df,
    category=["species", "island", "sex", "species", "island", "sex", "combine"],
    ncols=3,
    hatch=True,
)
```

```python
df = sns.load_dataset("titanic")
df = df[["class", "embark_town", "sex"]].fillna("NA")
res = op.nested_piechart(
    df,
    category=["class", "embark_town", "sex"],
    title="Titanic",
    ignore=0.01,
    show_legend=True,
)
```

```python
df = pd.read_csv("../data/owid-energy-data_curated_with_continents.csv")
df = df.loc[df["year"] == 2021]
res = op.nested_piechart(
    df,
    category=["Continent", "country"],
    variable="coal_electricity",
    title="Coal electricity in 2021 (terawatt-hours)",
    ignore=0.03,
    show_legend=True,
    show_values=True,
    show_percentage=True,
    order="largest",
    hatch=False,
    skip_na=True,
    colormode="hierarchical",
)
```
