```python
import seaborn as sns
from omniplot import plot as op
import pandas as pd
```

```python
df = pd.DataFrame(
    {
        "Experiments": ["exp1", "exp1", "exp1", "exp1", "exp2", "exp2", "exp3"],
        "GO": [
            "nucleus",
            "cytoplasm",
            "chromosome",
            "DNA binding",
            "chromosome",
            "RNA binding",
            "RNA binding",
        ],
        "FDR": [10, 1, 5, 3, 1, 2, 0.5],
        "odds": [3.3, 1.1, 2.5, 2.1, 0.8, 2.3, 0.9],
    }
)
op.dotplot(
    df,
    row="GO",
    col="Experiments",
    size_val="FDR",
    color_val="odds",
    highlight="FDR",
    color_title="Odds",
    size_title="-log10 p",
    scaling=20,
)
```
