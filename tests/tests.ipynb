```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import statannot
import omniplot.plot as op
```

```python
df = sns.load_dataset("penguins")
df = df.dropna(axis=0)
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
sns.violinplot(
    data=df, x="species", y="bill_depth_mm", inner="quartile", color="gray", ax=ax
)
sns.swarmplot(data=df, x="species", y="bill_depth_mm", color="black", size=3, ax=ax)
statannot.add_stat_annotation(
    ax,
    data=df,
    x="species",
    y="bill_depth_mm",
    box_pairs=[("Adelie", "Chinstrap"), ("Adelie", "Gentoo"), ("Chinstrap", "Gentoo")],
    test="Mann-Whitney",
    text_format="star",
    loc="inside",
    verbose=2,
)
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
op.violinplot2(
    df=df, x="species", y="bill_depth_mm", pairs=[["Adelie", "Gentoo"]], ax=ax
)
```
